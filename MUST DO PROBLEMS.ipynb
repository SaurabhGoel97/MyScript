```python
#install the following packages in your directory
#Step 1 open anaconda prompt
#step 2 activate the environment to install packages
#step 3 install the following packages by typing below commands

# pip install requests
# pip install html5lib
# pip install bs4


import requests 
import random
from bs4 import BeautifulSoup

URL = "https://www.geeksforgeeks.org/must-coding-questions-company-wise/"

r = requests.get(URL) 

soup = BeautifulSoup(r.content, 'html5lib') 

table = soup.findAll('li')

table2 = list()

random.seed(0)

random.shuffle(table)    #shuffling all the content in the list(problems)

for row in table:
    if row.parent.name=='ol':
        table2.append(row)

content = []

for i in table2[:-5]:
    cont = {}
    cont['url'] = i.a['href'] 
    cont['problem'] = i.text
    content.append(cont)
    

num = 1

file = open("D:\codingproblems.doc","a")  #specify the path to create file in your computer

file.write("MUST DO CODING PROBLEMS FOR INTERVIEW PREPARATION \n \n \n")

for row in content:
    a,b = row.values()
    s1 = str(num) + ")"+ " Problem Statement :" + b + "\n"
    file.write(s1)
    file.write
    s2 = "     " + "URL :" + "\n"
    file.write(s2)
    file.write(a)
    file.write("\n")
    num = num + 1
    

#run the following script to download the document containing all the must do problems
```
